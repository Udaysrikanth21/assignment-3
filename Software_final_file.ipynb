```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Rating Distribution with count values
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='Rating', data=df_cleaned, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
# Adding the text labels on the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.show()
```

```python
# Top 10 Most Reviewed Brands with annotations
plt.figure(figsize=(10, 6))
top_brands = df_cleaned['Brand Name'].value_counts().nlargest(10)
ax = sns.barplot(x=top_brands.index, y=top_brands.values, palette='coolwarm')
plt.title('Top 10 Most Reviewed Brands')
plt.xlabel('Brand Name')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
# Adding the text labels on the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.show()
```

```python
# Price vs. Rating with median annotations
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Rating', y='Price', data=df, palette='Set3')
plt.title('Price Distribution by Rating')
plt.xlabel('Rating')
plt.ylabel('Price ($)')
plt.ylim(0, 1000)  # Limiting y-axis for better visualization
# Annotating the median price inside each box
medians = df_cleaned.groupby(['Rating'])['Price'].median().values
median_labels = [f'${median:.2f}' for median in medians]
pos = range(len(medians))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 10, median_labels[tick],
            horizontalalignment='center', size='x-small', color='w', weight='semibold')

plt.show()
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset, DataLoader
```

```python
# Define the Dataset Class
class ReviewsDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
    def __len__(self):
        return len(self.texts)
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }
# Load the pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=5)  # Assuming 5 rating classes
```

```python
# Prepare the dataset
MAX_LEN = 128
BATCH_SIZE = 16
#  Convert ratings to numerical labels
df_cleaned['Rating'] = df_cleaned['Rating'].astype(int) - 1  # Assuming ratings are from 1 to 5, convert to 0-4 for labels
dataset = ReviewsDataset(
    texts=df_cleaned['Reviews'].tolist(),
    labels=df_cleaned['Rating'].tolist(),
    tokenizer=tokenizer,
    max_len=MAX_LEN
)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=1,  # Set to 1 for quick training
    per_device_train_batch_size=BATCH_SIZE,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)
# Trainer class for training the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)
```

```python
# Train the model
trainer.train()
```
